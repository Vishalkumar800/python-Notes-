# What is Variable

A variable is a name used to store data in memory. In Python, a variable is a reference to an object stored in memory. 
We use variables to store, access, and manipulate data throughout a program.

Python is a dynamically typed programming language. This means we do not need to explicitly declare the data type of a variable. The type is determined automatically at runtime.
